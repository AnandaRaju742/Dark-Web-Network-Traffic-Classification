Testing if Python Weka works


```python
import weka.core.jvm as jvm
jvm.start(packages=True,max_heap_size="25G")

from weka.core.converters import Loader
from weka.core.classes import Random
from weka.classifiers import Classifier, Evaluation, FilteredClassifier
from weka.filters import Filter
```

ARFF File loader

```python
loader = Loader(classname="weka.core.converters.ArffLoader")
data_file = 'Set 3 L2.arff'
data = loader.load_file(data_file)
print('Sample size: ', data.num_attributes)
print(data.attribute_names())
```

```python

cls = Classifier(classname="weka.classifiers.bayes.NaiveBayesMultinomial")
data.class_index = 0
from weka.filters import Filter
stw = Filter(classname="weka.filters.unsupervised.attribute.StringToWordVector", options=["-W", "10000","-C","-T","-N","1"])
from weka.classifiers import FilteredClassifier
fc = FilteredClassifier()
fc.filter=stw
fc.classifier = cls
#cls.build_classifier(data)
from weka.classifiers import Evaluation
from weka.core.classes import Random
evl = Evaluation(data)
evl.crossvalidate_model(fc, data, 10, Random(1))
#evl.crossvalidate_model(cls, data, 10, Random(1))

print("Accuracy: ",evl.percent_correct)
print("F-Measure: ",evl.unweighted_macro_f_measure)
print("\n")

```
